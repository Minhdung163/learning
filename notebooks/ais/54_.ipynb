## Intermezzo

Položka `agi_stub` definuje kategorii příjmů, odpovídající

    1 = 1 to 25,000
    2 = 25,000 to 50,000
    3 = 50,000 to 75,000
    4 = 75,000 to 100,000
    5 = 100,000 to 200,000
    6 = 200,000 or more

se středními hodnotami přibližně 

    1 = 12,500
    2 = 37,500
    3 = 62,500
    4 = 87,500
    5 = 112,500
    6 = 212,500


Položka `N1` počet subjektů s příjmy v dané kategorii.

## Příklad

### Zadání

Nahraďte kategorickou proměnnou `agi_stub` odhadem hodnoty.

```python
medians = {1: 12500, 2: 37500, 3: 62500, 4: 87500, 5: 112500, 6: 212500}
# df3["agi_stub"] = 
df3
```

### Řešení

```python

medians = {1: 12500, 2: 37500, 3: 62500, 4: 87500, 5: 112500, 6: 212500}
df3["agi_stub"] = df3.agi_stub.map(medians)
df3
```

## Příklad

### Zadání

Proveďte seskupení dat podle proměnné `zipcode` a následně spočítejte průměrný zisk, vytvořte a zobrazte dataframe

```python
df4 = None
# groups = df3.groupby(by="zipcode")
# df4 = pd.DataFrame(groups.)
df4
```

### Řešení

```python

groups = df3.groupby(by="zipcode")
groups["zipcode"]

df4 = pd.DataFrame(
    groups.apply(lambda x: sum(x["N1"] * x["agi_stub"]) / sum(x["N1"]))
)
df4
```
