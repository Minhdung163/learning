```python
import datetime
lowdate = datetime.datetime.fromisoformat("2023-08-01")
highdate = datetime.datetime.fromisoformat("2023-09-01")
ff = lambda item: lowdate <= item <= highdate
ii = list(map(ff, df["date1"]))
print(ii[:20])
df[ii]
```

```python
import datetime
def createtimedata():
    td = datetime.timedelta(minutes=1)
    tc = datetime.datetime.fromisoformat("2023-10-01")
    value = 1
    result = []
    for _ in range(24*60):
        item = {}
        item["date"] = tc
        item["value"] = value
        result.append(item)
        tc = tc + td
        value = value + 1
    return result
```

```python
import pandas as pd
df0 = pd.DataFrame(createtimedata())
df0
```

```python
df0.set_index("date", inplace=True)
df0
```

```python
data_Monthly = df0.resample('1M')
data_Monthly
```

```python
import pandas as pd
grouper = pd.Grouper(freq="1h")
df0['normed'] = df0.groupby(grouper).transform(lambda x: x/x.mean())
df0
```
